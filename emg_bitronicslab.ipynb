```python
%matplotlib inline
name="/content/data/Main/1-st.csv"
df=pandas.read_csv(name, sep=',')
print(df.head())
bigFingerTime=df['Time']
bigFingerTime=bigFingerTime-bigFingerTime[0]
bigFingerV=df['V']
print(bigFingerV.max())
```

```python
name="/content/data/Main/2-nd.csv"
df=pandas.read_csv(name, sep=",")
print(df.head())
luchTime=df['Time']
luchTime=luchTime-luchTime[0]
luchV=df['V']
print(luchV.max())
```

```python
name="/content/data/Main/3-rd.csv"
df=pandas.read_csv(name, sep=",")
print(df.head())
bicepsTime=df['Time']
bicepsTime=bicepsTime-bicepsTime[0]
bicepsV=df['V']
print(bicepsV.max())
```

```python
name="/content/data/Main/4-th.csv"
df=pandas.read_csv(name, sep=",")
print(df.head())
tricepsTime=df['Time']
tricepsTime=tricepsTime-tricepsTime[0]
tricepsV=df['V']
print(tricepsV.max())
```

```python
name="/content/data/Main/5-th.csv"
df=pandas.read_csv(name, sep=",")
print(df.head())
deltaTime=df['Time']
deltaTime=deltaTime-deltaTime[0]
deltaV=df['V']
print(deltaV.max())
```

```python
fig, ax = plt.subplots()
ax.plot(luchTime, luchV, "-", label='Luch')
ax.plot(bigFingerTime, bigFingerV, "-", label='Big Finger')  
ax.plot(bicepsTime, bicepsV, "-", label='Biceps') 
ax.plot(tricepsTime, tricepsV, "-", label='Triceps')
ax.plot(deltaTime, deltaV, "-", label='Delta')
ax.legend()
```

```python
luchPeaks=find_peaks(luchV, height = 4.35, distance = 20)
luchPeaks = luchPeaks[0]
print("Среднее значение напряжения при работе лучевой мышцы равно ", round(luchV[luchPeaks].mean(), 2))
plt.plot(luchTime, luchV)
plt.plot(luchTime[luchPeaks], luchV[luchPeaks], "ro")
plt.show()
```

```python
bigFingerPeaks=find_peaks(bigFingerV, height = 4.4, distance = 20)
bigFingerPeaks = bigFingerPeaks[0]
print("Среднее значение напряжения при работе большого пальца равно ", round(bigFingerV[bigFingerPeaks].mean(), 2))
plt.plot(bigFingerTime, bigFingerV)
plt.plot(bigFingerTime[bigFingerPeaks], bigFingerV[bigFingerPeaks], "ro")
plt.show()
```

```python
bicepsPeaks=find_peaks(bicepsV, height = 4.2, distance = 20)
bicepsPeaks = bicepsPeaks[0]
print("Среднее значение напряжения при работе бицепса равно ", round(bicepsV[bicepsPeaks].mean(), 2))
plt.plot(bicepsTime, bicepsV)
plt.plot(bicepsTime[bicepsPeaks], bicepsV[bicepsPeaks], "ro")
plt.show()
```

```python
tricepsPeaks=find_peaks(tricepsV, height = 4.7, distance = 20)
tricepsPeaks = tricepsPeaks[0]
print("Среднее значение напряжения при работе трицепса равно ", round(tricepsV[tricepsPeaks].mean(), 2))
plt.plot(tricepsTime, tricepsV)
plt.plot(tricepsTime[tricepsPeaks], tricepsV[tricepsPeaks], "ro")
plt.show()
```

```python
deltaPeaks=find_peaks(deltaV, height = 4.5, distance = 20)
deltaPeaks = deltaPeaks[0]
print("Среднее значение напряжения при работе дельтавидных мышц равно ", round(deltaV[deltaPeaks].mean(), 2))
plt.plot(deltaTime, deltaV)
plt.plot(deltaTime[deltaPeaks], deltaV[deltaPeaks], "ro")
plt.show()
```
